```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model as lm
```

```python
data = pd.read_csv('mtcars.csv')

correlation =  data.corr()                   
print(correlation)
```

```python
data.hist()
plt.show()
```

```python
x = data.drop(data.columns[[0,2]],axis = 1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9),color = "black")
```

```python
X = pd.DataFrame(data['wt'])
y = data['mpg']

model = lm.LinearRegression()
model.fit(X,y)
print(model.intercept_)
print(model.coef_)
model.score(X,y)
```

```python
y_pred = model.predict(X)
residual = data['mpg']-y_pred
residual.describe()
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9),color = "black")
plt.plot(data['wt'],y_pred,color='blue')
```

```python
# bac 2
model_2 = lm.LinearRegression()
A = pd.DataFrame([data['wt'],data['wt']**2]).T
b = data['mpg']
model_2.fit(A,b)

print('Intercept cua mo hinh bac hai:')
print(model_2.intercept_)
print('Coefficient cuar mo hinh bac hai')
print(model_2.coef_)
print('Do chinh xac cua mo hinh')
print(model_2.score(A,b))
```

```python
range = np.arange(1.5,5.5,0.1)
predictor = pd.DataFrame([range, range**2]).T
A = predictor
y_pred_2 = model_2.predict(A)
```

```python
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9),color = "black")
plt.plot(range,y_pred_2,color='blue')
```
